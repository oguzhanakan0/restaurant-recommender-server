```python
import pandas as pd
df = pd.read_csv("clean_data.csv")

def convertToFact(row):
    return f'restaurant("{row["name"]}","{row["cuisines"]}","{row["rest_type"]}",\
{row["rate"]},{row["dollar_sign"]},{row["restaurant_value"]},{row["x"]},{row["y"]}).\n'

df["asFact"] = df.apply(convertToFact, axis=1)
```

```python
df = df[df['name'].str.len()<40]
```

```python
with open('knowledgeBase.pl', mode="w") as file:
    file.writelines(df.asFact.tolist())
```
